Morphological transformations are some simple operations based on the image shape. It is normally performed on binary images.Two basic morphological operators are Erosion and Dilation. Then its variant forms like Opening, Closing.

`Erosion` : erodes away the boundaries of foreground object (Always try to keep foreground in white)

`Dilation` :  increases the white region in the image or size of foreground object increases

`Opening` : Erosion followed by dilation

`Closing` : Dilation followed by erosion

image

`Distance Transform` :  obtain the derived representation of a binary image, where the value of each pixel is replaced by its distance to the nearest background pixel

`Skeletonize` : reduces binary objects to 1 pixel wide representations
